```python
import ipywidgets as widgets
from IPython.display import display, clear_output

class QuizApp:
    def __init__(self):
        self.quiz_data = [
            {
                "question": "What is the capital of Japan?",
                "options": ["Tokyo", "Beijing", "Seoul", "Bangkok"],
                "correct_answer": 0
            },
            {
                "question": "Who wrote the novel 'Pride and Prejudice'?",
                "options": ["Jane Austen", "Emily Bronte", "Charlotte Bronte", "Louisa May Alcott"],
                "correct_answer": 0
            },
            {
                "question": "What is the chemical symbol for the element Gold?",
                "options": ["Ag", "Au", "Cu", "Fe"],
                "correct_answer": 1
            },
        ]
        self.current_question_index = 0
        self.score = 0

        # Widgets
        self.question_label = widgets.Label(value="")
        self.option_buttons = [widgets.Button(description="") for _ in range(4)]
        self.feedback_label = widgets.Label(value="")
        self.next_button = widgets.Button(description="Next Question")
        self.next_button.on_click(self.next_question)

        # Create Layout
        self.options_box = widgets.VBox(self.option_buttons)
        self.quiz_box = widgets.VBox([
            self.question_label,
            self.options_box,
            self.feedback_label,
            self.next_button
        ])

    def start_quiz(self):
        self.load_question()
        display(self.quiz_box)

    def load_question(self):
        question_data = self.quiz_data[self.current_question_index]
        self.question_label.value = question_data["question"]
        options = question_data["options"]
        self.feedback_label.value = ""
        for i, button in enumerate(self.option_buttons):
            button.description = options[i]
            button.on_click(lambda _, i=i: self.check_answer(i))
        self.next_button.disabled = True

    def check_answer(self, selected_option):
        question_data = self.quiz_data[self.current_question_index]
        correct_answer_index = question_data["correct_answer"]
        if selected_option == correct_answer_index:
            self.score += 1
            self.feedback_label.value = "✅ Correct!"
        else:
            self.feedback_label.value = "❌ Incorrect!"
        self.next_button.disabled = False
        for button in self.option_buttons:
            button.on_click(lambda _: None)  # Disable further clicks

    def next_question(self, _=None):
        self.current_question_index += 1
        if self.current_question_index == len(self.quiz_data):
            self.end_quiz()
        else:
            self.load_question()

    def end_quiz(self):
        clear_output()
        final_score_label = widgets.Label(value=f"Quiz Over! Your score: {self.score}/{len(self.quiz_data)}")
        display(final_score_label)

# Create and start the quiz
quiz_app = QuizApp()
quiz_app.start_quiz()
```
