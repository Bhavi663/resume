2.Convert image to HSV

3.Find contours( boundary of an image)

```python
import cv2
import matplotlib.pyplot as plt
# Load the original image
image = cv2.imread('tesla.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB for correct color display with matplotlib
# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# Save the grayscale image
cv2.imwrite('grayscale_image.jpg', gray_image)
# Display both images side by side
plt.figure(figsize=(10, 5))
# Original image
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
# Grayscale image
plt.subplot(1, 2, 2)
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
# Load the original image
image = cv2.imread('tesla.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB for correct color display with matplotlib
# Convert to HSV
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
# Save the HSV image
cv2.imwrite('hsv_image.jpg', hsv_image)
plt.imshow(hsv_image)
plt.title("HSV Image")
plt.show()
```

```python
# Load the image
image = cv2.imread('tesla.jpg')
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# Apply a binary threshold to get a binary image
_, threshold_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
# Find the contours
contours, _ = cv2.findContours(threshold_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
# Draw the contours on the original image
contour_image = cv2.drawContours(image.copy(), contours, -1, (0, 255, 0), 2)
# Save the image with contours
cv2.imwrite('contour_image.jpg', contour_image)
plt.imshow(contour_image)
plt.title("Contour Image")
plt.show()
```
