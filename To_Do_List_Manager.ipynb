```python
import ipywidgets as widgets
from IPython.display import display, clear_output

class TodoListManager:
    def __init__(self):
        self.tasks = []

        # Widgets for task list
        self.task_list = widgets.Select(options=[], rows=10, description="Tasks")

        # Input and buttons
        self.input_task = widgets.Text(placeholder="Enter a task", description="Task:")
        self.btn_add = widgets.Button(description="Add Task", button_style="success")
        self.btn_remove = widgets.Button(description="Remove Task", button_style="danger")

        # Display area for feedback
        self.feedback = widgets.Label()

        # Button event handlers
        self.btn_add.on_click(self.add_task)
        self.btn_remove.on_click(self.remove_task)

        # Layout
        self.layout = widgets.VBox([
            widgets.HBox([self.input_task, self.btn_add]),
            self.task_list,
            self.btn_remove,
            self.feedback
        ])

    def add_task(self, _):
        task = self.input_task.value.strip()
        if task:
            self.tasks.append(task)
            self.task_list.options = self.tasks
            self.input_task.value = ""
            self.feedback.value = "Task added successfully!"
        else:
            self.feedback.value = "Please enter a task."

    def remove_task(self, _):
        selected_task = self.task_list.value
        if selected_task:
            self.tasks.remove(selected_task)
            self.task_list.options = self.tasks
            self.feedback.value = f"Task '{selected_task}' removed successfully!"
        else:
            self.feedback.value = "No task selected."

    def run(self):
        display(self.layout)

# Create and run the To-Do List Manager
todo_list = TodoListManager()
todo_list.run()
```
