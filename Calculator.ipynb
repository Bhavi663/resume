```python
import ipywidgets as widgets
from IPython.display import display

# Function to evaluate the expression
def evaluate_expression(change):
    try:
        result = eval(expression.value)
        result_label.value = f"Result: {result}"
    except Exception as e:
        result_label.value = f"Error: {e}"

# Create the entry widget for the expression
expression = widgets.Text(value='', placeholder='Enter expression', description='Expression:', disabled=False)

# Create a label to display the result
result_label = widgets.Label(value="Result:")

# Create buttons for numbers and operators
button_1 = widgets.Button(description='1')
button_2 = widgets.Button(description='2')
button_3 = widgets.Button(description='3')
button_4 = widgets.Button(description='4')
button_5 = widgets.Button(description='5')
button_6 = widgets.Button(description='6')
button_7 = widgets.Button(description='7')
button_8 = widgets.Button(description='8')
button_9 = widgets.Button(description='9')
button_0 = widgets.Button(description='0')

button_add = widgets.Button(description='+')
button_sub = widgets.Button(description='-')
button_mul = widgets.Button(description='*')
button_div = widgets.Button(description='/')
button_clear = widgets.Button(description='C')
button_equal = widgets.Button(description='=')

# Function to add character to expression
def add_to_expression(button):
    current_expression = expression.value
    expression.value = current_expression + button.description

# Clear the expression
def clear_expression(button):
    expression.value = ''

# Set up button actions
button_1.on_click(add_to_expression)
button_2.on_click(add_to_expression)
button_3.on_click(add_to_expression)
button_4.on_click(add_to_expression)
button_5.on_click(add_to_expression)
button_6.on_click(add_to_expression)
button_7.on_click(add_to_expression)
button_8.on_click(add_to_expression)
button_9.on_click(add_to_expression)
button_0.on_click(add_to_expression)

button_add.on_click(add_to_expression)
button_sub.on_click(add_to_expression)
button_mul.on_click(add_to_expression)
button_div.on_click(add_to_expression)
button_clear.on_click(clear_expression)

button_equal.on_click(evaluate_expression)

# Display all widgets
display(expression, button_1, button_2, button_3, button_4, button_5, button_6, button_7, button_8, button_9, button_0,
        button_add, button_sub, button_mul, button_div, button_clear, button_equal, result_label)
```
