```python
# Step 1: Import required libraries
import cv2
from google.colab import files
from google.colab.patches import cv2_imshow

# Step 2: Upload a video file
uploaded = files.upload()

# Step 3: Load the pre-trained face detection cascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Step 4: Get the uploaded video file path
video_path = list(uploaded.keys())[0]

# Step 5: Initialize the video capture object
cap = cv2.VideoCapture(video_path)

# Step 6: Check if the video capture object is opened successfully
if not cap.isOpened():
    print("Error opening video capture")
    exit()

# Step 7: Process the video frames
while True:
    # Read the frame from the video capture
    ret, frame = cap.read()

    # Break the loop if the video ends
    if not ret:
        break

    # Step 8: Convert the frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Step 9: Detect faces in the grayscale frame
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # Step 10: Draw rectangles around the detected faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Step 11: Display the resulting frame using cv2_imshow in Colab
    cv2_imshow(frame)

    # Step 12: Wait a bit before showing the next frame
    cv2.waitKey(1)

# Step 13: Release the video capture object
cap.release()

# Step 14: Close any OpenCV windows (although cv2_imshow doesn't open traditional windows)
cv2.destroyAllWindows()
```
