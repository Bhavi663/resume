```python
import ipywidgets as widgets
from IPython.display import display, clear_output

class TicTacToe:
    def __init__(self):
        self.current_player = "X"
        self.board = [["" for _ in range(3)] for _ in range(3)]
        self.buttons = [[widgets.Button(description="", layout=widgets.Layout(width="100px", height="100px", font_size="24px")) for _ in range(3)] for _ in range(3)]
        self.status_label = widgets.Label(value="Player X's turn")
        self.grid = widgets.GridBox(
            children=[btn for row in self.buttons for btn in row],
            layout=widgets.Layout(grid_template_columns="repeat(3, 100px)", grid_gap="5px")
        )
        for i in range(3):
            for j in range(3):
                self.buttons[i][j].on_click(lambda _, i=i, j=j: self.make_move(i, j))
        self.game_box = widgets.VBox([self.status_label, self.grid])

    def make_move(self, row, col):
        if self.board[row][col] == "":
            self.board[row][col] = self.current_player
            self.buttons[row][col].description = self.current_player
            if self.check_winner(self.current_player):
                self.status_label.value = f"Player {self.current_player} wins!"
                self.disable_buttons()
            elif self.is_board_full():
                self.status_label.value = "It's a draw!"
                self.disable_buttons()
            else:
                self.current_player = "O" if self.current_player == "X" else "X"
                self.status_label.value = f"Player {self.current_player}'s turn"

    def check_winner(self, player):
        for i in range(3):
            if all(self.board[i][j] == player for j in range(3)) or all(self.board[j][i] == player for j in range(3)):
                return True
        if self.board[0][0] == self.board[1][1] == self.board[2][2] == player or self.board[0][2] == self.board[1][1] == self.board[2][0] == player:
            return True
        return False

    def is_board_full(self):
        return all(self.board[i][j] != "" for i in range(3) for j in range(3))

    def disable_buttons(self):
        for row in self.buttons:
            for btn in row:
                btn.on_click(lambda _: None)

    def start_game(self):
        display(self.game_box)

# Start the Tic Tac Toe game
tic_tac_toe = TicTacToe()
tic_tac_toe.start_game()
```
