```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import cv2  # To read the image
import numpy as np
image = cv2.imread('flag.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for correct color representation
# Split the image into R, G, B channels
channels = ('Red', 'Green', 'Blue')
colors = ('r', 'g', 'b')
plt.figure(figsize=(12, 6))
# Calculate and plot histogram for each channel
for i, color in enumerate(colors):
  histogram = cv2.calcHist([image], [i], None, [256], [0, 256])
  sns.lineplot(x=range(256), y=histogram.flatten(), color=color, label=f'{channels[i]} Channel')
plt.title('Color Distribution Histogram')
plt.xlabel('Intensity Value')
plt.ylabel('Pixel Count')
plt.legend()
plt.show()
```
